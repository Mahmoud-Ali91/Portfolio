```python
import os
import pandas as pd

# Folder where all your CSVs are stored
folder_path = "output_tables"  # Update this to your actual folder

# Load and combine
all_data = []
for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(folder_path, file))
        df['Source_File'] = file  # Add a column to track original file
        all_data.append(df)

# Concatenate all
combined_df = pd.concat(all_data, ignore_index=True)

# Save to a new CSV
combined_df.to_csv("Combined_Aesthetic_Market.csv", index=False)

print("✅ All CSVs combined into: Combined_Aesthetic_Market.csv")
```
